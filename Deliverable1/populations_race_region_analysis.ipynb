```python
import pandas as pd
import matplotlib.pyplot as plt

df_neighborhood = pd.read_csv("./Boston_neighborhood_data.csv")
df_neighborhood = df_neighborhood.iloc[:, :6]
df_neighborhood.columns = df_neighborhood.iloc[0]
df_neighborhood = df_neighborhood.iloc[1:]
df_neighborhood.set_index('field concept', inplace = True)
df_neighborhood.index.name = None
df_neighborhood[['Total:', 'White alone', 'Black or African American alone', 'Hispanic or Latino', 'Asian, Native Hawaiian and Pacific Islander alone, all ages']] = df_neighborhood[['Total:', 'White alone', 'Black or African American alone', 'Hispanic or Latino', 'Asian, Native Hawaiian and Pacific Islander alone, all ages']].apply(pd.to_numeric)
df_neighborhood
```

```python
ax = df_neighborhood.plot(y=['White alone', 'Black or African American alone', 'Hispanic or Latino', 'Asian, Native Hawaiian and Pacific Islander alone, all ages'], kind="bar", stacked = True, title="Population based on race and region")
plt.legend(list(df_neighborhood.columns.values)[1:])
horiz_offset = 1.03
vert_offset = 1.
ax.legend(bbox_to_anchor=(horiz_offset, vert_offset))
plt.show()
```
